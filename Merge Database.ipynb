### Location

```python
location1 = pd.read_csv("location1.csv") 
location2 = pd.read_csv("location2.csv") 
location3 = pd.read_csv("location3.csv")
location4 = pd.read_csv("location4.csv")
location5 = pd.read_csv("location5.csv")
location6 = pd.read_csv("location6.csv")
```

```python
location = pd.concat([location1, location2, location3, location4, location5, location6], ignore_index=True)
```

```python
location.drop_duplicates(keep='last', inplace=True)
location = location.drop(columns=['IsReadOnly', 'Latitude', 'Longitude', 'Elevation', 'Width', 'Depth', 'Height', 'IsPartSource', 'IsPartDestination', 'IsZoneBatchPickZone', 'IsZoneWavePickZone', 'IsPickZoneAutoStart', 'IsWorkstation', 'ParentLocationId', 'PickZoneTemplate'])
location = location.reset_index(drop=True)
```

```python
location["ID"].duplicated().sum()
```

```python
location
```

```python
location['CreatedDate'] =pd.to_datetime(location.CreatedDate)
location.CreatedDate = location.CreatedDate.dt.tz_localize('GMT')
location.CreatedDate = location.CreatedDate.dt.tz_convert('US/Eastern')
location.CreatedDate = location.CreatedDate.dt.tz_localize(None)
```

```python
location['UpdatedDate'] =pd.to_datetime(location.UpdatedDate)
location.UpdatedDate = location.UpdatedDate.dt.tz_localize('GMT')
location.UpdatedDate = location.UpdatedDate.dt.tz_convert('US/Eastern')
location.UpdatedDate = location.UpdatedDate.dt.tz_localize(None)
```

```python
location['DeletedDate'] =pd.to_datetime(location.DeletedDate)
location.DeletedDate = location.DeletedDate.dt.tz_localize('GMT')
location.DeletedDate = location.DeletedDate.dt.tz_convert('US/Eastern')
location.DeletedDate = location.DeletedDate.dt.tz_localize(None)
```
